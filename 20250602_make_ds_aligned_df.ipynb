```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pkl = pd.read_pickle('ds_aligned_spikes.pkl')
```

```python
sample_id = pd.RangeIndex(2500, name='sample')
condition_ids = pd.RangeIndex(4, name='condition')
session_ids = pd.Series(pkl.keys(), name='session')
```

```python
list3 = []
for session_id in session_ids:
    cell_ids = pd.Series(pkl[session_id]['cell_id'], name='cell')
    spike_counts = pkl[session_id]['spike_count']
    ds_combo = pkl[session_id]['ds_combo']
    list2 = []
    for condition_id in condition_ids:
        list1 = []
        for i, cell_id in enumerate(cell_ids):
            combo_mask = ds_combo == condition_id
            # presentation_id = pd.RangeIndex(combo_mask.sum(), name='presentation')
            ds_isi = pd.Series(pkl[session_id]['ds_isi'][combo_mask], name='isi')
            array = pkl[session_id]['ds_psth'][i, combo_mask, :]
            df = pd.DataFrame(array, index=ds_isi, columns=sample_id)
            list1.append(df)
        df1 = pd.concat(list1, axis=0, keys=cell_ids)
        list2.append(df1)
    df2 = pd.concat(list2, axis=0, keys=condition_ids)
    list3.append(df2)
result = pd.concat(list3, axis=0, keys=session_ids)
result = result.reorder_levels(['condition', 'session', 'isi', 'cell'])
result = result.sort_index()
```

```python
pd.to_pickle(result, 'ds_aligned_spikes_df.pkl')
```
